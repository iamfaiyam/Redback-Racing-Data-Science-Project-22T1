```python
import numpy as np 
import pandas as pd
```

```python
data0_00_output = pd.read_csv('~/Downloads/data0/data0_00/output.csv')
data0_00_input = pd.read_csv('~/Downloads/data0/data0_00/input_params.csv')
```

```python
# Need to aggregate the output data first 
data = data0_00_output
df = data.groupby('acsys.CS.LapCount').agg(['mean', 'max', 'min'])
```

```python
# Transposing the input data to match with output 
data0_00_input.set_index('parameter', inplace = True) 
data0_00_input = data0_00_input.transpose()
```

```python
# Replicate the rows of input to match number of rows of output data 
data0_00_input = data0_00_input.append([data0_00_input]*14, ignore_index = True)
```

```python
# Append replicated input -> output data 
df = df.join(data0_00_input)
```

```python
#retrieve a list of all the data files from data0_00 to data0_50 
data_fold_list = []

i = 0
start = 'data0_'
while i <= 50:
    data_fold = start
    
    if i <= 9: #00, 01, 02, ..., 09
        data_fold = data_fold + '0' + str(i)
        data_fold_list.append(data_fold)
        
    else: #10, 11, 12, ..., 50
        data_fold = data_fold + str(i)
        data_fold_list.append(data_fold) 
    
    i = i + 1

print(data_fold_list) # list of all data folders
```

```python
# mass importing of all data files (output and input) for each data folder
data0_output = [] # data0_00_o, data0_01_o, ..., data0_50_o
data0_input = [] # data0_00_o, data0_01_i, ..., data0_50_i

i = 0
front = '~/Downloads/data0/'
backi = '/input_params.csv'
backo = '/output.csv'
while i <= 50:
    curro = front
    curri = front 
    
    curro = curro + data_fold_list[i] + backo 
    curri = curri + data_fold_list[i] + backi
    
    data0_output.append(curro) 
    data0_input.append(curri) 
    i = i + 1
```

```python
i = 0 
lap_count = []
while i <= 50: 
    data0_out_curr = pd.read_csv(data0_output[i])
    data0_inp_curr = pd.read_csv(data0_input[i])
    
    if i == 0: # base case
        
        # aggregating the output data 
        df = data.groupby('acsys.CS.LapCount').agg(['mean', 'max', 'min'])
        lap_count.append(df.shape[0] - 1)
        
        # drop last row of each lap due to potential premature cutoff
        df.drop(index = df.index[-1], axis = 0, inplace = True) 
        
        # reshaping input data 
        data0_inp_curr.set_index('parameter', inplace = True) 
        data0_inp_curr = data0_inp_curr.transpose() 
        
        # replicating rows of input to match number of rows of output 
        data0_inp_curr = data0_inp_curr.append([data0_inp_curr]*lap_count[i],ignore_index = True)
        
        # appending replicated input data to output data 
        df = df.join(data0_inp_curr) 
        
    else: #multiple case 
        
        # aggregating the output data 
        df2 = data.groupby('acsys.CS.LapCount').agg(['mean', 'max', 'min'])
        lap_count.append(df2.shape[0] - 1)
        
        # drop last row of each lap due to potential premature cutoff
        df2.drop(index = df2.index[-1], axis = 0, inplace = True) 
        
        # reshaping input data 
        data0_inp_curr.set_index('parameter', inplace = True) 
        data0_inp_curr = data0_inp_curr.transpose()
        
        # replicating rows of input to match number of rows of output 
        data0_inp_curr = data0_inp_curr.append([data0_inp_curr]*lap_count[i], ignore_index = True) 
        
        # appending replicated intput data to output data 
        df2 = df2.join(data0_inp_curr) 
        
        df = df.append(df2, ignore_index = True) 
    
    i = i + 1
```
